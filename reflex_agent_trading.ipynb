# Reflex + Agentic AI Trading System (Indian Market)

This notebook demonstrates:
- Reflex trading agent (rule-based)
- Support & Resistance detection
- Chart visualization
- Agent-style structured output

```

import yfinance as yf
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```

def get_data(symbol="^NSEI", period="5d", interval="5m"):
    df = yf.download(symbol, period=period, interval=interval)
    df = df.reset_index()
    return df

df = get_data()
df.tail()
```

```

def support_resistance(df):
    supports, resistances = [], []
    for i in range(2, len(df)-2):
        if df['Low'][i] < df['Low'][i-1] and df['Low'][i] < df['Low'][i+1]:
            supports.append(df['Low'][i])
        if df['High'][i] > df['High'][i-1] and df['High'][i] > df['High'][i+1]:
            resistances.append(df['High'][i])
    return supports[-3:], resistances[-3:]

supports, resistances = support_resistance(df)
supports, resistances
```

```

def reflex_agent(price, supports, resistances):
    if price <= min(supports):
        return "BUY"
    elif price >= max(resistances):
        return "SELL"
    return "HOLD"

current_price = df['Close'].iloc[-1]
signal = reflex_agent(current_price, supports, resistances)
signal
```

```

fig = go.Figure()

fig.add_trace(go.Candlestick(
    x=df['Datetime'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']
))

for s in supports:
    fig.add_hline(y=s)

for r in resistances:
    fig.add_hline(y=r)

fig.show()
```

```

output = {
    "symbol": "NIFTY",
    "trend": "auto",
    "support_levels": supports,
    "resistance_levels": resistances,
    "signal": signal
}

output
```
